```python
# Exercise 1: prime number
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):  
        if n % i == 0:
            return False
    return True


number = int(input("Enter a number: "))

if is_prime(number):
    print(number, "is a prime number.")
else:
    print(number, "is not a prime number.")
```

```python
#   Exercise 2: Product of Random Numbers

import random

def multiplication_quiz():
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    num3 = num1 * num2

    # Ask the user for their answer
    try:
        num = int(input(f"What is {num1} x {num2}? "))

        # Check the answer
        if num == num3:
            print("Correct! Well done.")
        else:
            print(f"Incorrect. The correct answer is {num}.")
    except ValueError:
        print("Invalid input. Please enter an integer.")
        
multiplication_quiz()
```

```python
#   Exercise 3: Squares of Even/Odd Numbers
for num in range(100, 202):  # Choosing even number
    if num % 2 == 0:
        square = num ** 2
        print(f"{num} squared is {square}")
```

```python
#Exercise 4: Problem Statement (Question): Hospital Billing System

Room_charges_per_day = 2000
Docter_fee = 1500
Lab_test_cost = 300
Bill = 10000
Discount_rate = 0.10

days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400

room_charges = Room_charges_per_day * days_stayed
lab_charges = Lab_test_cost * num_lab_tests
doctor_fee = Docter_fee

Total_before_discount = room_charges + doctor_fee + lab_charges + medicine_charges

if Total_before_discount > Bill:
    discount = Total_before_discount * Discount_rate
else:
    discount = 0

Final_bill = Total_before_discount - discount

print(" Hospital Billing system ")
print(f"Room charges (₹{Room_charges_per_day}/day x {days_stayed} days): ₹{room_charges}")
print(f"Doctor consultation fee: ₹{Docter_fee}")
print(f"Lab test charges (₹{Lab_test_cost} x {num_lab_tests} tests): ₹{lab_charges}")
print(f"Medicine charges: ₹{medicine_charges}")
print(f"Total before discount: ₹{Total_before_discount}")
print(f"Discount applied: ₹{discount}")
print(f"Final bill : ₹{Final_bill}")
```
