```python
# 1.Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

numbers = [10, 20, 30, 40, 50]
print("Original list:", numbers)

# append
numbers.append(60)
print("After append(60):", numbers)

# extend
numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)

# insert
numbers.insert(2, 25)  # inserts 25 at index 2
print("After insert(2, 25):", numbers)

# remove
numbers.remove(40)
print("After remove(40):", numbers)

# pop
popped_element = numbers.pop()
print("After pop():", numbers)
print("Popped element:", popped_element)

# index
index_of_30 = numbers.index(30)
print("Index of 30:", index_of_30)

# count
count_of_20 = numbers.count(20)
print("Count of 20:", count_of_20)

# sort
numbers.sort()
print("After sort():", numbers)

# reverse
numbers.reverse()
print("After reverse():", numbers)

# clear (at the very end)
numbers.clear()
print("After clear():", numbers)
```

```python
# 7. Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

num = int(input("enter value: "))
if num > 0:
    if num % 2 == 0:
        print("Positive and Even")
    else:
        print("Positive but Odd")
elif num == 0:
    print("Zero")
else:
    print("Negative")
```

```python
# 11. Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
# - Use .info(), .describe()
# - Select 'Name' and 'City' columns
# - Drop 'City' column
# - Fill any missing values in 'Salary' column with the mean
# - Remove any duplicate rows

import pandas as pd
import numpy as np

data = {
    'Name': ['Sandesh', 'Raju', 'Sahil', 'Ravi', 'Priyanka', 'Ganesh', 'Anand', 'Veda', 'Raghavendra', 'Pallavi'],
    'Age': [15, 25, 25, 22, 27, 29, 24, 27, 26, 30],
    'City': ['Bangalore', 'Hyderabad', 'Pune', 'Delhi', 'Bangalore', 'Delhi', 'Pune', 'Mumbai', 'Delhi', 'Bangalore'],
    'Salary': [35000,50000, 60000, 55000, np.nan, 60000, 45000, 52000, 60000, np.nan]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n", df, "\n")

print("Info about DataFrame:")
print(df.info(), "\n")

print("Statistical Summary (.describe()):")
print(df.describe(), "\n")

print("Selected Columns (Name & City):")
print(df[['Name', 'City']], "\n")

df_dropped = df.drop(columns=['City'])
print("After dropping 'City' column:\n", df_dropped, "\n")

mean_salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(mean_salary)
print("After filling missing Salary values:\n", df, "\n")

df_no_duplicates = df.drop_duplicates().reset_index(drop=True)
print("After removing duplicate rows:\n", df_no_duplicates)
```
