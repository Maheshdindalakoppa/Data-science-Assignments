```python
import numpy as np
import scipy.stats as stats
```

```python
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 
                 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
n = len(data)
mean = np.mean(data)
s = np.std(data, ddof=1)
```

```python
 # Confidence level
confidence = 0.99
alpha = 1 - confidence

# Degrees of freedom
df = n - 1

# t critical value for 99% CI
t_critical = stats.t.ppf(1 - alpha/2, df)

# Standard Error
SE_t = s / np.sqrt(n)

# Margin of Error
ME_t = t_critical * SE_t

# Confidence Interval
CI_t_lower = mean - ME_t
CI_t_upper = mean + ME_t
```

```python
print("A) 99% Confidence Interval using sample standard deviation (t-distribution):")
print(f"Mean = {mean:.3f}")
print(f"Sample Std Dev (s) = {s:.3f}")
print(f"t-critical = {t_critical:.3f}")
print(f"Confidence Interval = ({CI_t_lower:.3f}, {CI_t_upper:.3f})")
print("-" * 60)
```

```python
# Given population standard deviation
sigma = 0.2

# z critical value for 99% CI
z_critical = stats.norm.ppf(1 - alpha/2)

# Standard Error
SE_z = sigma / np.sqrt(n)

# Margin of Error
ME_z = z_critical * SE_z

# Confidence Interval
CI_z_lower = mean - ME_z
CI_z_upper = mean + ME_z
```

```python
print("B) 99% Confidence Interval using known population standard deviation (z-distribution):")
print(f"Mean = {mean:.3f}")
print(f"Population Std Dev (σ) = {sigma:.3f}")
print(f"z-critical = {z_critical:.3f}")
print(f"Confidence Interval = ({CI_z_lower:.3f}, {CI_z_upper:.3f})")
```

```python
import numpy as np
import scipy.stats as stats
```

```python
# Inputs
X_mean, sigma_X = 600, 25  # mean number of units produced in a week ,standard deviation of units
sample_mean, n, alpha = 3050, 25, 0.05  # sample mean weekly cost , sample size , significance level

# 1. Parameter Calculation
# Calculate theoretical mean weekly cost (μ)
mu = 1000 + (5 * X_mean)
print(f"Theoretical Mean Weekly Cost (μ) = Rs. {mu:.3f}")

# Calculate population standard deviation of weekly cost (σ) ,Since W = 1000 + 5X, the standard deviation of W = 5 * σ_X
sigma = 5 * sigma_X
print(f"Population Standard Deviation of Weekly Cost (σ) = Rs. {sigma:.3f}")

# Calculate the Z Test Statistic
standard_error = sigma / np.sqrt(n)

# 2. Statistical Calculation
# We use the Normal distribution directly to find the p-value

# Calculate the Z Test Statistic
z = (sample_mean - mu) / standard_error

# Determine the critical value and p-value
p_value = stats.norm.cdf(z) # sf is the Survival Function (1 - cdf)
z_critical = stats.norm.ppf(1 - alpha)

# 3. Output Results
print(f"Z-Statistic: {z:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Critical Value: {z_critical:.4f}")

if p_value < alpha:
    print("\n Conclusion")
    print("Decision: Reject H0 - There is strong evidence at the 5% significance level that the average weekly operating cost.is higher than the cost predicted by the model.")
else:
    print("Decision: There is insufficient evidence to conclude that the average weekly operating cost is higher than the cost predicted by the model. ")
```

```python
 # (W = 1000 + 5X): Defines the linear relationship where weekly cost (W)) is a fixed 1000 base plus 5 times the units produced (X).

# Theoretical Mean: The average cost (Rs.4000) calculated by plugging the average production level (\bar={X}=600) into the model.

# Cost Spread: The population standard deviation of cost (5 \times \sigma_X = 125), representing the expected fluctuation in weekly spending.

# The hypothesis  test : checks whether the true mean cost exceeds Rs. 4000.
# Null Hypothesis (H_{0}:\  mu=4000)
# Alternative Hypothesis (H_{1}:\ mu>4000)

# Conclusion : static test  Z
# Since the (Z) score is too low, the data does not provide enough evidence to prove a cost increase; the original model is still valid.  
```
